```python
import os
import yaml
import pandas as pd
from pandas import json_normalize
from yaml import safe_load
from tqdm import tqdm
```

```python
# Extract File Names
folder_path = 'data'

file_names = [f for f in os.listdir(folder_path) if f.endswith('.yaml')]

print(f"Total files found: {len(file_names)}")
print("First 5 files:", file_names[:5])
```

```python
# to extract yaml files 
final_df = pd.DataFrame()
counter = 1

for file_name in tqdm(file_names, desc="Loading match files"):
    full_path = os.path.join(folder_path, file_name)
    
    with open(full_path, 'r') as f:
        df = pd.json_normalize(safe_load(f), max_level=1)
        df['match_id'] = counter
        final_df = pd.concat([final_df, df], ignore_index=True)
        counter += 1

final_df
```

```python
# relevant columns
final_df = final_df[[
    'match_id',
    'info.dates',
    'info.venue',
    'info.city',
    'info.gender',
    'info.match_type',
    'info.overs',
    'info.teams',
    'info.umpires',
    'info.toss',
    'info.outcome',
    'innings',
]].rename(columns={
    'info.dates'      : 'date',
    'info.venue'      : 'venue',
    'info.city'       : 'city',
    'info.gender'     : 'gender',
    'info.match_type' : 'match_type',
    'info.overs'      : 'overs',
    'info.teams'      : 'teams',
    'info.umpires'    : 'umpires',
    'info.toss'       : 'toss',
    'info.outcome'    : 'outcome',
})
```

```python
import pickle
df = pickle.load(open('dataset_level1.pkl', 'rb'))
df.head()
```

```python
all_rows = []

for _, row in tqdm(df.iterrows(), desc="Extracting ball by ball", total=len(df)):

    innings = row['innings']

    # We only want the 1st innings
    if not innings or len(innings) == 0:
        continue

    first_innings = innings[0]  # 0 = 1st innings, 1 = 2nd innings

    for inning_name, inning_content in first_innings.items():

        batting_team = inning_content.get('team', None)
        deliveries   = inning_content.get('deliveries', [])

        # Figure out bowling team
        teams        = row['teams']
        bowling_team = [t for t in teams if t != batting_team][0] if teams else None

        cumulative_runs    = 0
        cumulative_wickets = 0

        for delivery in deliveries:
            for ball, ball_data in delivery.items():

                runs_total          = ball_data.get('runs', {}).get('total', 0)
                cumulative_runs    += runs_total
                cumulative_wickets += 1 if ball_data.get('wicket') else 0

                all_rows.append({
                    'match_id'           : row['match_id'],
                    'date'               : row['date'],
                    'venue'              : row['venue'],
                    'city'               : row['city'],
                    'gender'             : row['gender'],
                    'toss_winner'        : row['toss'].get('winner', None),
                    'toss_decision'      : row['toss'].get('decision', None),
                    'winner'             : row['outcome'].get('winner', None),
                    'batting_team'       : batting_team,
                    'bowling_team'       : bowling_team,
                    'ball'               : ball,
                    'runs_batsman'       : ball_data.get('runs', {}).get('batsman', 0),
                    'runs_extras'        : ball_data.get('runs', {}).get('extras', 0),
                    'runs_total'         : runs_total,
                    'wicket'             : 1 if ball_data.get('wicket') else 0,
                    'cumulative_runs'    : cumulative_runs,
                    'cumulative_wickets' : cumulative_wickets,
                })

# Convert to DataFrame
ball_by_ball_df = pd.DataFrame(all_rows)

print(f"Shape: {ball_by_ball_df.shape}")
ball_by_ball_df.head(10)
```
