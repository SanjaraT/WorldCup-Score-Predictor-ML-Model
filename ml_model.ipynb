```python
import os
import yaml
import pandas as pd
from pandas import json_normalize
from yaml import safe_load
from tqdm import tqdm
```

```python
# Extract File Names
folder_path = 'data'

file_names = [f for f in os.listdir(folder_path) if f.endswith('.yaml')]

print(f"Total files found: {len(file_names)}")
print("First 5 files:", file_names[:5])
```

```python
# to extract yaml files 
final_df = pd.DataFrame()
counter = 1

for file_name in tqdm(file_names, desc="Loading match files"):
    full_path = os.path.join(folder_path, file_name)
    
    with open(full_path, 'r') as f:
        df = pd.json_normalize(safe_load(f), max_level=1)
        df['match_id'] = counter
        final_df = pd.concat([final_df, df], ignore_index=True)
        counter += 1

final_df
```

```python
import pickle
df = pickle.load(open('dataset_level1.pkl', 'rb'))
df.head()
```

```python
# Extract ball by ball data for every match's first innings

# for each match
all_rows = []

for index, row in tqdm(df.iterrows(), desc="Extracting deliveries", total=len(df)):

    ball_of_match      = []
    batsman            = []
    bowler             = []
    runs               = []
    player_dismissed   = []
    teams              = []
    batting_team       = []
    match_id           = []
    city               = []
    venue              = []

    try:
        deliveries = row['innings'][0]['1st innings']['deliveries']
    except:
        continue  

    for ball in deliveries:
        for key in ball.keys():
            match_id.append(row['match_id'])
            batting_team.append(row['innings'][0]['1st innings']['team'])
            teams.append(row['teams'])
            ball_of_match.append(key)
            batsman.append(ball[key]['batsman'])
            bowler.append(ball[key]['bowler'])
            runs.append(ball[key]['runs']['total'])
            city.append(row['city'])
            venue.append(row['venue'])

            # Get dismissed player if wicket fell, else '0'
            try:
                player_dismissed.append(ball[key]['wicket']['player_out'])
            except:
                player_dismissed.append('0')

    loop_df = pd.DataFrame({
        'match_id'         : match_id,
        'teams'            : teams,
        'batting_team'     : batting_team,
        'ball'             : ball_of_match,
        'batsman'          : batsman,
        'bowler'           : bowler,
        'runs'             : runs,
        'player_dismissed' : player_dismissed,
        'city'             : city,
        'venue'            : venue,
    })

    all_rows.append(loop_df)


delivery_df = pd.concat(all_rows, ignore_index=True)

print(f"Shape: {delivery_df.shape}")
delivery_df.head(10)
```

```python
delivery_df['bowling_team'] = delivery_df.apply(
    lambda row: [team for team in row['teams'] if team != row['batting_team']][0], 
    axis=1
)

```

```python
import pickle
matches = pickle.load(open('dataset_level2.pkl', 'rb'))
matches.tail(10)
```

```python
matches.isnull().sum()
```

```python
final_cities = matches['city'].value_counts() [matches['city'].value_counts()>600].index.tolist()
    
```

```python
matches = matches[matches['city'].isin(final_cities)]
```

```python
matches
```

```python
# current score
matches['current_score'] = matches.groupby('match_id')['runs'].cumsum()
```

```python
matches
```

```python
# Left balls
# spliting balls ( 1.1 -> 1 over 1 ball)
matches['over'] = matches['ball'].apply(lambda x: int(str(x).split('.')[0]))
matches['ball_in_over'] = matches['ball'].apply(lambda x: int(str(x).split('.')[1]))

# total balls bowled so far
matches['balls_bowled'] = matches['over'] * 6 + matches['ball_in_over'].apply(lambda x : min(x,6))

# Balls left = 120 - balls bowled
matches['balls_left'] = 120 - matches['balls_bowled'].clip(0, 120)

matches
```

```python
# wickets left
matches['player_dismissed'] = matches['player_dismissed'].apply(lambda x:0 if x=='0' else 1)
matches['player_dismissed'] = matches['player_dismissed'].astype('int')
matches['player_dismissed'] = matches.groupby('match_id')['player_dismissed'].cumsum()
matches['wickets_left'] = 10 - matches['player_dismissed']
```

```python
matches
```

```python
matches['current_run_rate'] = (
    matches['current_score'] / (matches['balls_bowled'] / 6)
).replace([float('inf'), float('nan')], 0).round(2)

matches
```
