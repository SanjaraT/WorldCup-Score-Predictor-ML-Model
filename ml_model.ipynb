```python
import os
import yaml
import pandas as pd
from pandas import json_normalize
from tqdm import tqdm
```

```python
# Extract File Names
folder_path = 'data'

file_names = [f for f in os.listdir(folder_path) if f.endswith('.yaml')]

print(f"Total files found: {len(file_names)}")
print("First 5 files:", file_names[:5])
```

```python
all_data = []

# tqdm for wraping the loop and showing a progress bar
for file_name in tqdm(file_names, desc="Loading match files"):
    full_path = os.path.join(folder_path, file_name)

    # load yaml data
    with open(full_path, 'r') as f:
        match_data = yaml.safe_load(f) 

    # to store in table format
    flat_data = json_normalize(match_data['info'])
    flat_data['file_name'] = file_name

    all_data.append(flat_data)

final_df = pd.concat(all_data, ignore_index=True)
final_df
```
